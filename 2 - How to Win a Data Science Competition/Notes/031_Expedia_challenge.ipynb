# Week 1: Expedia Kaggle Competition (Video part 28)

- Problem: Predicting which hotel group a user will book based on user's search results, clicks, bookings and information about travel package.
- Prediction target is hotel group - characteristics of actual hotel.


### Data leakage in the competition
- `destinaton_distance` - distance from user city to actual hotel a user clicked and booked.
- Our prediction target is a characteristic of an actual hotel.
- Since `destination_distance` is very precise, we could use the unique `destination_distance` and `user_city` pairs is a leakage to true hotel location.
- This works perfectly when we have such pair present in train set, we could directly use it as a label for test set.
- But nearly half of the rows in test set are new pairs, so we need to create new features from the destination_distance and user_city pair.
- For example, how many hotels of which group there are for user city, hotel country etc.
