### Goal:
- Different validation strategies
- How many splits should we make?
- What methods to perform splits?


### Validation types
- Holdout
- k-folds
- leave-one-out
- The main difference between the above validation strategies is the number of splits being made.


#### Holdout
- Simplest technique: split data into train and test.
- Train/test sets must not overlap - which might be the case when we have duplicates rows. So make sure to drop duplicates and confirm that our train/test sets do not overlap.


#### K-folds
- Core idea of k-fold is to use every subsample for validation once.
- Good method when we have small data.



#### Leave-one-out
- If k = len of samples in our data, then it will iterate through every sample in the data



#### Stratification
- We should use stratification to ensure all subsample will get similar portion of target label, otherwise our model will be biased to the target value it sees most.
- Important for: small datasets, imbalanced dataset, multiclass classification

# Summary
- If we have enough data, and we're likely to get similar scores and optimal model parameter for different splits, use holdout.
- If optimal parameters is different for different splits, use K-fold.
- If we have small dataset, use leave-one-out.
- Use stratification to ensure the model see all target labels equally.
