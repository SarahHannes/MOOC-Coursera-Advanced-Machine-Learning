# Week 1: Leaderboard probing and examples of rare data leaks (Video part 27)


- 2 types of leaderboard probing:
    - Extracting all ground truth from public leaderboard.
    - Extracting information from private leaderboard.
- Target label could have the same distribution for public and private leaderboard


- Adapted log loss for leaderboard probing. Where:
    - C = constant prediction
    - N = number of rows
    - N1 = number of rows with target of 1
    - L = leaderboard score given by that constant prediction.
- From this formula, we can calc N1 divided by N ie true ratio of target=1 in the test set.
- Using this information, we can re-balance the training data points to have the same distribution of target variable in the test set.
