# Week 2: Exploring anonymized data (Video part 19)

## Objective:
- What is anonymized data?
- What can we do with it?

### What is anonymized data?
- Data that the organizers intentionally changed.


### What can we do with it?
- Try to de-anonymize the data (guess the meaning)
- Guess the types of columns (separating them into categorical, numeric)
- Try to find how the features are related to each other

- Useful functions:
  - df.dtypes: 3 feature types (object=can be anything (eg numeric with nans, or has some text), float, integer=can be categorical/bool/numeric)
  - df.info()
  - x.value_counts(): check value counts
  - x.isnull()
