### Competition tips:
- 2 main reasons some people jump down on leaderboard after private results is revealed:
  - Competitors ignore the validation rules and choose submission which scored best against the public leaderboard.
  - The competition has no consistent private/public data split. Or too little data in either private/public leaderboard.


### Goal:
- Systematic ways to set up validation for a competition
- Tackle most common validation problems



### Validation
- We want to check if the model give expected results on unseen data.
- We want to make sure the model will be applicable in the future, and its quality
- In competitions, test data is split into public set and private set. When we send submissions, we see the score for public test set but scores for private test set is only released at the end.
- If we got very good scores on the public test set but got bad score on private test set, then we have overfitted model.



- Overfitting in general = validation result on train set is better than test set
- Overfitting in competitions = low model quality on test dat, which was unexpected due to validation scores
