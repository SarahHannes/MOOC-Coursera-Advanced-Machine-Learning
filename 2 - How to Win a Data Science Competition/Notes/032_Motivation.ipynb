### Metrics
- Metrics are used to evaluate our submissions, and there are plenty of ways to measure equality.
- Take away point: if your model is scored with some metric, you get best result by optimizing exactly that metric.
- Some metrics can not be optimized efficently, eg if there is no simple enough way to find the optimal hyperplane. This is why sometimes we need to train our model to optimize something different than competition metric. But in this case, we need to apply various heuristics to improve competition metric score.
- Another reason why we sometimes need to optimize metric different than the competition metric is when the train and test sets are different.
- In leaderboard probing, we check if the mean target value on the public part of test set is the same as on train - if not, we would need to adapt our predictions to suit test set better.
- In severe case, improved metric on the validation does not result in improved metric on the test set. In this case, we need to approach the problem differently.
- We need to understand what the competition metric wants and try to design optimization metric to reach that goal if it is harder to optimize over the original competition metric:
  - In the below case study, the competition metric's goal is to correctly predict trend than to predict the actual value. So instead of training model to predict values, we train model to predict (ylast + some tiny number = ie uptrend) or (ylast - some tiny number = ie downtrend).
